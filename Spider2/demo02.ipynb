```python
from selenium import webdriver
import time
from selenium.webdriver import ActionChains
from PIL import Image
from io import BytesIO

path = "D:\\the_geckodriver\\geckodriver.exe"

firefox = webdriver.Firefox(executable_path=path)

# 控制火狐浏览器访问目标网站
firefox.get("https://zhuanlan.zhihu.com/p/88319091")

# 窗口最大化（方便后续截图）
firefox.maximize_window()

time.sleep(2)

# 找到弹出登录窗口的账号密码验证元素
change_Identity_Way_ele = firefox.find_element_by_xpath(
    "/html[1]/body[1]/div[4]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/form[1]/div[1]/div[2]")

# 点击切换到账号密码验证
change_Identity_Way_ele.click()

time.sleep(1)

# 自己知乎网站的账号密码
phone_number = "15608685519"
password = ".3xLnQe7MPY2sc5"

# 找到账号密码输入框
phone_numbers_ele = firefox.find_element_by_xpath(
    "/html[1]/body[1]/div[4]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/form[1]/div[2]/div[1]/label[1]/input[1]")
password_ele = firefox.find_element_by_xpath(
    "/html[1]/body[1]/div[4]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/form[1]/div[3]/div[1]/label[1]/input[1]")

# 将自己的账号密码填入输入框
phone_numbers_ele.send_keys(phone_number)
password_ele.send_keys(password)

time.sleep(2)

# 找到登录按钮
login_ele = firefox.find_element_by_xpath(
    "/html[1]/body[1]/div[4]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/form[1]/button[1]")

# 点一次不行,就多点他丫的几次
try:
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
    login_ele.click()
except Exception as e:
    print("麻了呀!点多了,出现：", e)

time.sleep(2)

# 鼠标移动到验证窗口的滑块上
try:
    slider_btn = firefox.find_element_by_class_name(
        "yidun_slider__icon")

    ActionChains(firefox).move_to_element(slider_btn).perform()
except Exception as e:
    print("还没点击到登录呜呜呜呜哭了!", e)


time.sleep(2)

# 先截取整个界面图片


def crop_image(image_file_name=""):
    # 定位某个元素在浏览器中的位置
    img_ele = firefox.find_element_by_xpath(
        "/html[1]/body[1]/div[6]/div[2]/div[1]/div[1]/div[2]/div[1]/div[1]/div[1]/div[1]")

    # 获取图片元素的起始位置(以左上角为坐标原点)
    img_location = img_ele.location

    print(img_location)

    # 获取图片的尺寸
    img_size = img_ele.size
    print(img_size)

    # 算出整个图片四个顶点的位置
    top, buttom, left, right = img_location["y"], img_location["y"] + img_size["height"],img_location["x"],img_location["x"] + img_size["width"]
        
    screenshot = firefox.get_screenshot_as_png()
    screenshot = Image.open(BytesIO(screenshot))
    captcha = screenshot.crop((int(left),int(top),int(right),int(buttom)))
    captcha.save(image_file_name)
    return captcha


crop_image("image1.png")



time.sleep(10)
```
