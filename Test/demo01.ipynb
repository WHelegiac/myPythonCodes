```python
import pandas as pd
from collections import Counter

data = pd.read_excel("./newdata.xlsx")

counter = Counter(list(data["sex"]))

x = ["男性","女性"]
y = [counter[1],counter[2]]

pie = Pie().add('',[list(z) for z in zip(x,y)]).set_series_opts(label_opts=opts.LabelOpts(formatter="{b}: {d}%"))

pie.render_notebook()
```

```python
x = [''.join(random.sample(string.ascii_letters,5)) for _ in range(10)]
y = [random.randint(0,100000) for _ in range(10)]


bar = Bar().add_xaxis(x).add_yaxis('',y)

pie = Pie().add('',[list(z) for z in zip(x,y)])


grid = (Grid()
        .add(bar,grid_opts=opts.GridOpts(pos_bottom="65%", pos_left="50%"))
        .add(pie,grid_opts=opts.GridOpts(pos_left="15%"))
        )


grid.render_notebook()
```
