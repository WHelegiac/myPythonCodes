```python
import random
import matplotlib.pyplot as plt
import math
from collections import Counter

def toss():
    return random.choice([0,1])


def do_experiment(n = 10000):
    counter = Counter([toss() for _ in range(n)])
    return counter[1]/n

x = [index for index in range(1,1000)]
y = [do_experiment(n) for n in x]

plt.plot(x,y)
```
