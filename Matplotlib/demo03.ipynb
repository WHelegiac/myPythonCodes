```python
import matplotlib.pyplot as plt 
import random 
```

```python
heights = [random.randint(160,190) for _ in range(100)]

bins = range(160,191,3)
plt.hist(heights,bins=bins)
plt.show()
```

```python

import matplotlib.pyplot as plt

years = range(1999,2024)
heights = [random.randint(160,190) for _ in range(25)]

plt.plot(years,heights)

plt.show()
```

```python
from matplotlib import pyplot as plt
import numpy as np
#添加图形对象
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
#使得X/Y轴的间距相等
ax.axis('equal')
#准备数据
langs = ['C', 'C++', 'Java', 'Python', 'PHP']
students = [23,17,35,29,12]
#绘制饼状图
ax.pie(students, labels = langs,autopct='%1.2f%%')
plt.show()
```
